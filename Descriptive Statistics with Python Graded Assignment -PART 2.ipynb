```python
#Quantitative & continuous variables
Quan_Var=['Age','Experience','Income' ,'CCAvg','Mortgage']
```

```python
# Measures of central Tendency -----> Mean,Median

Mean_val = df1[Quan_Var].mean()
Median_val = df1[Quan_Var].median()

print("Mean :")
print(Mean_val)
print('*'*30)
print("Median :")
print(Median_val)
print('*'*30)
```

```python
# Measures Of dispersion -- Range,Variance,Standard Deviation

min_val =df1[Quan_Var].min()
max_val =df1[Quan_Var].max()

print("Minimum Value : ")
print(min_val)
print('-'*40)
print("Maximum Value : ")
print(max_val)
print('-'*40)
Range = max_val- min_val
print("Range : ")
print(Range)
```

```python
# Co-efficient of range
coeff_range = ((max_val-min_val)/(max_val+min_val))*100
print(coeff_range)
```

```python
Variance_val = df1[Quan_Var].var()
print("Variance Of Quantitative Variables are : ")
print(Variance_val)
```

```python
Std_Dev = df1[Quan_Var].std()
print(" Standard Deviation Of Quantitative Variables are : ")
print(Std_Dev)
```

```python
### Co-efficient of variation  ----- std/mean

coeff_var = (Std_Dev/Mean_val)*100

print(coeff_var)
```

```python
# 19 :  How many customers use online banking? Do customers using bank internet facilities have higher income?

#Solution:

Total_online_customers = df1['Online'].sum()

# Calculate the average income for customers using bank internet facilities
Avg_income_online_banking = df1.loc[df1['Online'] == 1, 'Income'].mean()

# Calculate the average income for customers not using bank internet facilities
Avg_income_offline_banking = df1.loc[df1['Online'] == 0, 'Income'].mean()

print("Number of customers using online banking:", Total_online_customers)
print("Average income for customers using bank Online facilities:", Avg_income_online_banking)
print("Average income for customers not using bank Online facilities:", Avg_income_offline_banking)


# Interpretation : It appears that customers using bank Online facilities have a slightly higher average income
#                  compared to those who do not use bank internet facilities. 
```

```python
# 20. Using the z-score of the income variable, find out the number of observations outside the +-3Ïƒ.

# Solution: 

Mean_Income = df1['Income'].mean()
Std_Dev_Income = df1['Income'].std()

# Z SCORE :

df1['Z_Score'] = (df1['Income'] - Mean_Income) / Std_Dev_Income

# Total observations outside +-3 standard deviations

Total_outside_3std = ((df1['Z_Score'] > 3) | (df1['Z_Score'] < -3)).sum()

print("Number of observations outside +-3 standard deviations:", Total_outside_3std)
```
